4.Есть ли статистически значимые различия в росте дочерей?
Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165
Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160


```python
import numpy as np
from scipy import stats
```

```python
# Рост матерей и дочерей
heights_mothers = [172, 177, 158, 170, 178, 175, 164, 160, 169, 165]
heights_daughters = [173, 175, 162, 174, 175, 168, 155, 170, 160]
```

```python
# Выполняем t-тест
t_statistic, p_value = stats.ttest_ind(heights_mothers, heights_daughters)
```

```python
# Определяем уровень значимости
alpha = 0.05
```

```python
# Сравниваем p-значение с уровнем значимости
if p_value < alpha:
    print("Статистически значимые различия в росте дочерей есть.")
else:
    print("Статистически значимых различий в росте дочерей нет.")
```
